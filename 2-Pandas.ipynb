8-Her kıta için ve aşağıdaki alkol cinsleri için belirtilen hesaplamaları gerçekleştiriniz.

beer_servings - std
spirit_servings - median 
wine_servings - min, mean, max
total_litres_of_pure_alcohol - mean, count

```python
import numpy as np
```

```python
#count nasıl yapcağım en sonda?
df.groupby("continent").aggregate({"beer_servings": [np.std],"spirit_servings": [np.median],"wine_servings":[min,np.mean,max],"total_litres_of_pure_alcohol":[np.mean,'count']})
```

10- Yeni değişkenler üreterek her alkol cincisin ilgili ülkedeki toplam alkol tüketimi içerisindeki oranlarını yazdırınız.  

```python
df.head()
```

```python
b=df.drop(["continent","total_litres_of_pure_alcohol"],axis=1)
b["Rate_Beer"]=df["beer_servings"]/(df["beer_servings"]+df["spirit_servings"]+df["wine_servings"])
b["Rate_Spirit"]=df["spirit_servings"]/(df["beer_servings"]+df["spirit_servings"]+df["wine_servings"])
b["Rate_wine"]=df["wine_servings"]/(df["beer_servings"]+df["spirit_servings"]+df["wine_servings"])
b.columns=["country","beer_servings","Rate_Beer","spirit_servings","Rate_Spirit","wine_servings","Rate_wine"]
b
```
